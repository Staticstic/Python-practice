# K-means clustering 1

## 1단계: Centroid 배치

### k 값을 정하고 k만큼 centroid 배치

## 2단계: 클러스터 형성

### 생성한 centroid와 각 데이터 사이의 거리 계산하여 가까운 곳으로 클러스터 생성

## 3단계: centroid 위치 갱신

### 생성한 클러스터 별 데이터들의 평균값을 사용하여 중심을 찾고 중심으로 centroid 갱신

## 4단계: 클러스터 재형성

### 새롭게 갱신한 centroid 기준으로 각각의 데이터의 거리를 다시 계산하여 클러스터 생성

## 5단계: centroid 위치 갱신

### 새롭게 형성된 클러스터를 중심으로 centroid 갱신

### centroid의 위치를 더이상 갱신하지 않을 때까지 반복
