## 최적의 k 값 선정

적절한 k 값을 선정해야 데이터의 특성에 맞게 클러스터를 잘 나눌 수 있다.<br>
최적의 클러스터의 개수는 클러스터 별로 속한 데이터와 centroid의 거리 합이 최소가 되는 k가 될수 있다.<br>

### Inertia<br>
Inertia: 각 클러스터에 속한 데이터들과 centroid 사이의 거리들의 제곱합<br>

### Elbow Method<br>
inertia는 명확한 기준이 없어 판단하기가 어렵다. 따라서 서로 다른 k-means 모델을 만들어 inertia 값을 비교해봐야 한다.<br>

```Python
# inertia 값 저장할 리스트
inertias = []     

for k in range(1, 16): # k값의 범위 1~15로 지정
    model = KMeans(n_clusters=k, random_state=123)
    model.fit(scaled_df)
    inertias.append(model.inertia_)

# k값에 따른 inertia값 시각화
sns.lineplot(x=range(1, 16), y=inertias, marker='o')
```

위의 그래프는 k가 4~6 구간에서 기울기가 급격하게 줄어드는 것을 확인할 수 있다. <br>
이럴 때는 보통 k를 4, 5, 6으로 각각 지정하여 클러스터링을 해본 후 해석하기 좋은 k를 최적의 개수로 지정한다.<br>
이렇게 k가 증가함에 따라 inertia가 줄어드는 정도가 급격하게 작아지는 지점을 찾으면 최적 클러스터 개수를 정하는 것을 **Elbow Method**라고 부른다.
